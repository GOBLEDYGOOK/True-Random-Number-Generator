```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
with open('randomNumbers100k.txt') as f:
    lines = f.read().splitlines()

x=[]
for line in lines:
    x.append(int(line))

n, bin_edges = np.histogram(x, bins = 256,range=(0,255))
totalcount = np.sum(n)
bin_probability = n / totalcount
bins_per_point = np.fmin(np.digitize(x, bin_edges), len(bin_edges)-1)
probability_perpoint = [bin_probability[bins_per_point[i]-1] for i in range(len(x))] 

plt.figure(figsize=(16,8))
plt.xlabel("Value (xi)")
plt.ylim(top=0.006)
plt.ylabel("Probability (pi)")
plt.title("Empirical distribution of probability")
plt.hist(x=x, range=(0,255), density=True, bins=256)

entropy = - np.sum([(y * math.log(y,2)) for y in bin_probability])
print("Entropy: "+str(entropy))
```
